```
import mediapipe as mp
mp_pose = mp.solutions.pose

help(mp_pose.Pose)

# Initialize MediaPipe Pose.
pose = mp_pose.Pose(
    static_image_mode=True, min_detection_confidence=0.5)
```

```
# Prepare DrawingSpec for drawing the face landmarks later.
mp_drawing = mp.solutions.drawing_utils 
drawing_spec = mp_drawing.DrawingSpec(thickness=1, circle_radius=1)
```

```
for name, image in images.items():
  # Convert the BGR image to RGB and process it with MediaPipe Pose.
  results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  # Print nose landmark.
  image_hight, image_width, _ = image.shape

  if not results.pose_landmarks:
    continue
  print(f'coordinates: ('
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y * image_hight}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y * image_hight},'
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * image_hight}'
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x * image_width}, '
    f'{results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y * image_hight})'
  )
  ##angles
  
  # Draw pose landmarks.
  print(f'Pose landmarks of {name}:')
  #np.savetxt('test.out', name, delimiter=',')
  annotated_image = image.copy()
  mp_drawing.draw_landmarks(
      image=annotated_image,
      landmark_list=results.pose_landmarks,
      connections=mp_pose.POSE_CONNECTIONS,
      landmark_drawing_spec=drawing_spec,
      connection_drawing_spec=drawing_spec)
  #cv2_imshow(annotated_image)
  path = '/content/sample_data/Sorted_Dolphin_Plank_pose' 
  cv2.imwrite(os.path.join(path , name), annotated_image)
  #cv2.imwrite('name.jpg', annotated_image)
  #f=open('file.txt', 'w')
  #for name, annotated_image in images.items():
     # print(name)  
     # path = '/content/akarna1/aks' 
     # cv2.imwrite(os.path.join(path , name), annotated_image)
      #cv2_imshow(image)
      #cv2.waitKey(0)
 
  #for name, annotated_image in images.items():
      #path = '/content/data3' 
      #cv2.imwrite(os.path.join(path , name), annotated_image)
      #cv2_imshow(annotated_image)
  #cv2.waitKey(0)
  

```

```
import numpy as np
for name, image in images.items():
    results = pose.process(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  # Print nose landmark.
    image_hight, image_width, _ = image.shape
    if not results.pose_landmarks:
       continue
    #######RHip
    a = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y])
    b = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y])
    c = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y])
    ########Rrightknee 
    d = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y])
    e = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_KNEE].y])
    f = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ANKLE].y])
    #######rightelbow 
    g = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y])
    h = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y])
    i = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y])
    ########Rshoulder
    x = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y])
    y = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y])
    z = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_HIP].y])
    ###########LHip
    j = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y])
    k = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y])
    l = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y])
    ##############Lefttknee
    m = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y])
    n = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_KNEE].y])
    o = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ANKLE].y])
    ###############Leftelbow
    p = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y])
    q = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y])
    r = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y])
    ######################
    ########Lshoulder
    s = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y])
    t = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y])
    v = np.array([results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].x,results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_HIP].y])
    ######################
    ba = a - b
    bc = c - b
    cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
    angle = np.arccos(cosine_angle)
    print(f'RHip angles of {name}:')
    print (np.degrees(angle))
    ############
    ed = d-e
    ef = f-e
    cosine_angle = np.dot(ed, ef) / (np.linalg.norm(ed) * np.linalg.norm(ef))
    angle = np.arccos(cosine_angle)
    print(f'Rknee angles of {name}:')
    print (np.degrees(angle))
    ##############
    hg = g-h
    hi = i-h

    cosine_angle = np.dot(hg, hi) / (np.linalg.norm(hg) * np.linalg.norm(hi))
    angle = np.arccos(cosine_angle)
    print(f'RElbow angles of {name}:')
    print (np.degrees(angle))
    ###############
    kj = j-k
    kl = l-k

    cosine_angle = np.dot(kj, kl) / (np.linalg.norm(kj) * np.linalg.norm(kl))
    angle = np.arccos(cosine_angle)
    print(f'Lhip angles of {name}:')
    print (np.degrees(angle))
    #################
    nm = n-m
    no = o-n
    cosine_angle = np.dot(nm, no) / (np.linalg.norm(nm) * np.linalg.norm(no))
    angle = np.arccos(cosine_angle)
    print(f'LKnee angles of {name}:')
    print (np.degrees(angle))
    ###############
    qp = p-q
    qr = r-q
    cosine_angle = np.dot(qp, qr) / (np.linalg.norm(qp) * np.linalg.norm(qr))
    angle = np.arccos(cosine_angle)
    print(f'LElbow angles of {name}:')
    print (np.degrees(angle))
    ################

    yx = x-y
    yz = z-y
    cosine_angle = np.dot(yx, yz) / (np.linalg.norm(yx) * np.linalg.norm(yz))
    angle = np.arccos(cosine_angle)
    print(f'Rshoulder angles of {name}:')
    print (np.degrees(angle))
    ##############

    ts = s-t
    tv = v-t
    cosine_angle = np.dot(ts, tv) / (np.linalg.norm(ts) * np.linalg.norm(tv))
    angle = np.arccos(cosine_angle)
    print(f'Lshoulder angles of {name}:')
    print (np.degrees(angle))
    ##############
#print(a,b,c)
```
